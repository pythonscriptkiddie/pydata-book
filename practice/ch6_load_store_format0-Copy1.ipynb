pd.read_csv opens comma-separated-files like examples.com/ex1.csv

```python
df = pd.read_csv('examples/ex1.csv')
df
```

```python
#files do not always have a row for headers
df = pd.read_csv('examples/ex1.csv', header=None)
df
```

```python
#we can specify names ourselves
df = pd.read_csv('examples/ex1.csv', names=['a', 'b', 'c', 'd', 'message'])
df
```

```python
#index_col : specifies that we want a particular column as the index
df = pd.read_csv('examples/ex1.csv', index_col='message')
df
```

```python
#index_col : specifies that we want a particular column as the index
df = pd.read_csv('examples/ex1.csv', index_col=4)
df
```
