```python
#either axis can have a heirarchical index
pd.DataFrame(np.arange(12))
```

```python
#either axis can have a heirarchical index
pd.DataFrame(np.arange(12).reshape((4, 3)))
```

```python
#either axis can have a heirarchical index
pd.DataFrame(np.arange(12).reshape((4, 3)),
            index=[['a', 'a', 'b', 'b'],[1,2,1,2]])
```

```python
#either axis can have a heirarchical index
pd.DataFrame(np.arange(12).reshape((4, 3)),
            index=[['a', 'a', 'b', 'b'],[1,2,1,2]],
            columns=[['Ohio', 'Ohio', 'Colorado'],
                    ['Green', 'Red', 'Green']])
```

```python
#either axis can have a heirarchical index
frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
            index=[['a', 'a', 'b', 'b'],[1,2,1,2]],
            columns=[['Ohio', 'Ohio', 'Colorado'],
                    ['Green', 'Red', 'Green']])
```

```python
frame
```

```python
#heirarchical levels can have names
frame.index.names
```

```python
#heirarchical levels can have names
frame.index.names = ['key1', 'key2']
```

```python
frame.columns.names = ['state', 'color']
```

```python
frame.index.nlevels
```

```python
frame
```

```python
frame['Ohio']
```

```python
#create multindex from scratch
pd.MultiIndex.from_arrays([['Ohio', 'Ohio', 'Colorado'],
                         ['Green', 'Red', 'Green']],
                         names=['state', 'color'])
```

```python
#reordering and sorting levels
#swap levels on axis or sort data by one specific level
#.swaplevel method: Take two level numbers or names and swap them
#new object with levels interchanged, data unaltered
frame.swaplevel('key1', 'key2')
```

```python
frame.swaplevel('key2', 'key1') #same as above
```

```python
frame.swaplevel(0, 1)
```

```python
#same as above
frame.swaplevel(1, 0)
```

```python
#sort_index by default sorts the data lexicographically using all the index levels, but you 
#can use only one of the levels
frame.sort_index()
```

```python
frame.sort_index(level=1)
```

```python
#let's check out level 0
frame.sort_index(level=0)
```

```python
frame.swaplevel(0, 1).sort_index(level=0)
```

```python
#same as above
frame.swaplevel(1, 0).sort_index(level=0)
```

```python
#summary statistics by level
#often we have a level option for summary statistics
frame.groupby(level='key2')
```

```python
#summary statistics by level
#often we have a level option for summary statistics
frame.groupby(level='key2').sum()
```

```python
frame.groupby(level='color', axis='columns').sum()
```

```python
#now let's do frame.groupby(level='key2')
frame.groupby(level='key2', axis='rows').sum()
```

```python
frame.groupby(level='key1', axis='rows').sum()
```

```python
#indexing with a DataFrame's columns
pd.DataFrame({'a': range(7)})
```

```python
pd.DataFrame({'a': range(7),
             'b': range(7, 0, -1)})
```

```python
pd.DataFrame({'a': range(7),
             'b': range(7, 0, -1),
             'c': ['one', 'one', 'one', 'two', 'two',
                  'two', 'two']})
```

```python
pd.DataFrame({'a': range(7),
             'b': range(7, 0, -1),
             'c': ['one', 'one', 'one', 'two', 'two',
                  'two', 'two'],
             'd': [0, 1, 2, 0, 1, 2, 3]})
```

```python
frame = pd.DataFrame({'a': range(7),
             'b': range(7, 0, -1),
             'c': ['one', 'one', 'one', 'two', 'two',
                  'two', 'two'],
             'd': [0, 1, 2, 0, 1, 2, 3]})
```

```python
frame
```

```python
#df.set_index() creates a new DataFrame using one or more of the columns as the index.
frame2 = frame.set_index(['c', 'd'])
```

```python
frame2
```

```python
#by default, pandas removes the columns from the DataFrame, but they can be retained
#by passing the drop=False argument to df.set_index()
frame.set_index(['c', 'd'], drop=False)
```

```python
#by default, pandas removes the columns from the DataFrame, but they can be retained
#by passing the drop=False argument to df.set_index()
frame.set_index(['c', 'd'], drop=True) #redundant, because it is dropped by default
```

```python
#by default, pandas removes the columns from the DataFrame, but they can be retained
#by passing the drop=False argument to df.set_index()
frame.set_index(['d', 'c'], drop=True) #redundant, because it is dropped by default
```

```python
#by default, pandas removes the columns from the DataFrame, but they can be retained
#by passing the drop=False argument to df.set_index()
frame.set_index(['c', 'd'], drop=True).sum(axis=0) #redundant, because it is dropped by default
```

```python
frame.set_index(['c', 'd'], drop=True).sum(axis=1)
```

```python
#reset_index does the opposite of set_index, with the heirarchical index levels in the columns
frame2.reset_index()
```
