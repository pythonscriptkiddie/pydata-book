```python
import numpy as np
import pandas as pd
```

```python
#Sometimes, we have to create a PeriodIndex from arrays. For example, we have in the following
#macrodataset a year and quarter in different columns.
data = pd.read_csv('examples/macrodata.csv')
```

```python
data.head(5)
```

```python
data['year']
```

```python
data['quarter']
```

```python
#With these arrays and a frequency, we can combine them to form an index for this DataFrame.
#We can thus replace the numeric index with a DatetimeIndex.
index = pd.PeriodIndex(year=data['year'], quarter=data['quarter'],
                      freq='Q-DEC')
```

```python
index
```

```python
data.index = index
```

```python
data
```

```python
data['1959':'1960']
```

```python
data['1959-01-03': '1960-01-03'] #we can pass dates like with any other index
```
