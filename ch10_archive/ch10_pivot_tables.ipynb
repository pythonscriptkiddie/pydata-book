```python
import numpy as np
import pandas as pd
```

```python
tips0 = pd.read_csv('examples/tips.csv')
```

```python
tips0.head()
```

```python
tip_pct = tips0['tip']/ (tips0['total_bill'] - tips0['tip'])
```

```python
tip_pct
```

```python
tips = tips0.assign(tip_pct=tip_pct)
```

```python
tips
```

```python
tips.pivot_table(index=['day', 'smoker'])
```

```python
#to replace this behavior
tips_two = tips.reindex(pd.Index(['total_bill', 'tip', 'smoker', 'day', 'size', 'tip_pct'], dtype='object'), axis=1)
```

```python
tips_two.pivot_table(index=['day', 'smoker'])
```

```python
#we can also directly produce the same pivot table with groupby
tips.groupby(['day', 'smoker']).mean()
```

```python
#we can also directly produce the same pivot table with groupby
tips.groupby(['day', 'smoker']).mean(numeric_only=True)
```

```python
#what if we wanted to take the average of only tip_pct and size
#and additionally group by time
#smoker in table columns
#time and day in the rows
tips.pivot_table(index=['time', 'day'], columns='smoker',
                values=['tip_pct', 'size'])
```

```python
#include partial totals by passing margins=True
#this adds ALL row and column labels, with the corresponding values as
#the tier's group statistics
tips.pivot_table(index=['time', 'day'], columns='smoker',
                values=['tip_pct', 'size'], margins=True)
#Below: the "all" values do not take into account smoker status, the "Yes" and "No" values do
```

```python
#using aggregation functions other than mean for a pivot table
#use the aggfunc argument
tips.pivot_table(index=['time', 'smoker'], columns='day',
                values='tip_pct', aggfunc=len, margins=True) #includes null values in group sizes
```

```python
#using aggregation functions other than mean for a pivot table
#use the aggfunc argument
tips.pivot_table(index=['time', 'smoker'], columns='day',
                values='tip_pct', aggfunc='count', margins=True) #excludes null values in group sizes
```

```python
#if some combinations are empty, we can use a fill_value
tips.pivot_table(index=['time', 'size', 'smoker'],
                columns='day', values='tip_pct',
                fill_value=0)
```

```python
#crosstabs: used for cross-tabulations, and comp[uting roup frequencies
#data = #! ipython id=c4ea7c59067e4ed5acce365201cec661
from io import StringIO
#! blockstart
data = """Sample  Nationality  Handedness
1   USA  Right-handed
2   Japan    Left-handed
3   USA  Right-handed
4   Japan    Right-handed
5   Japan    Left-handed
6   Japan    Right-handed
7   USA  Right-handed
8   USA  Left-handed
9   Japan    Right-handed
10  USA  Right-handed"""
#! blockend
data = pd.read_table(StringIO(data), sep="\s+")
```

```python
data
```

```python
#we can summarize the data with pd.crosstab (more convenient than pivot table)
pd.crosstab(data['Nationality'], data['Handedness'], margins=True)
```

```python
data.pivot_table(index=['Nationality'], columns='Handedness',
                values='Sample', aggfunc='count', margins=True) #same as the above
```

```python
#first two argumenmts to crosstab: can be an array or list of arrays
pd.crosstab([tips['time'], tips['day']], tips['smoker'], margins=True)
```
